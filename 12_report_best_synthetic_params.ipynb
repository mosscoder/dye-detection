```python
import pickle
import optuna
from tabulate import tabulate
```

```python
focus = 'augs'
results_dir = f'/content/drive/MyDrive/Colab_Notebooks/dye_test_opt/ternary/results/{focus}'
synthetic_study_path = results_dir.replace(f'{focus}', 'synthetic_properties/study.pkl')

with open(synthetic_study_path, 'rb') as f:
    synthetic_study = pickle.load(f)

params = synthetic_study.best_trial.params
```

```python
params
```

```python
# Manually format the LaTeX table with bold headers
latex_table_with_caption = r"""
\begin{table}[ht]
\centering
\begin{tabular}{ll}
\hline
\textbf{Parameter}       & \textbf{Value}   \\
\hline
"""

# Add rows to the table by iterating through the dictionary
for key, value in params.items():
    latex_table_with_caption += f"{key.replace('_', ' ').title()} & {value} \\\\\n"

# Close the table structure and add caption and label
latex_table_with_caption += r"""
\hline
\end{tabular}
\caption{Results from Bayesian optimization of color properties of synthetic dye overlays}
\label{tab:bayesian_optimization_results}
\end{table}
"""

# Print or save the LaTeX table output with caption
print(latex_table_with_caption)
```
